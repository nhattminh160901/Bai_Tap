```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 10000
sap = []
ngua = []
p_sap = []
p_ngua = []
# Hàm để tính số lần xuất hiện mặt ngửa

for _ in range(M):
  if random.random() < 0.5:
    ngua.append(1)
    sap.append(0)
  else:
    ngua.append(0)
    sap.append(1)

p_sap = [sum(sap[:i])/(sum(sap[:i])+sum(ngua[:i])) for i in range(1, M+1)]
p_ngua = [sum(ngua[:i])/(sum(sap[:i])+sum(ngua[:i])) for i in range(1, M+1)]
```

```python
plt.plot(p_sap)
plt.plot(p_ngua)
```

```python
import random
import matplotlib.pyplot as plt

# Số lần tung xúc xắc
M = 10000
# Khởi tạo danh sách để lưu trữ xác suất của mỗi mặt sau mỗi lần tung
probabilities_over_time = {i: [] for i in range(1, 7)}


# Tung xúc xắc và cập nhật xác suất sau mỗi lần tung
results = []
for _ in range(M):
    roll = random.randint(1, 6)
    results.append(roll)
    for i in range(1, 7):
        probabilities_over_time[i].append(results.count(i) / len(results))
```

```python
probabilities_over_time[3]
```

```python
plt.figure(figsize=(20,6))
for i in range(1, 7):
    plt.plot(probabilities_over_time[i], label=f'Mặt {i}')
plt.xlabel('Số lần tung')
plt.ylabel('Xác suất')
plt.legend()
plt.show()
```

```python
# height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
# weight (kg)
y = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
```

```python
lr = 0.0000025
T = 2
w = np.zeros(T)
N = X.shape[0]
for t in range(1,T):
  tmpw = 0.0
  for i in range(N):
    tmpw += lr*(y[1]-w[t]*X[i])*X[i]
  w[t] = w[t-1] + tmpw/N

print(w)
```

```python
record = np.zeros(3).tolist()
```

```python
for _ in range(100):
  N = np.random.randint(2, size=3).tolist()
  posR = []
  while True:
    pos = random.randint(0, len(N)-1)
    if pos not in posR:
      posR.append(pos)
      if N[pos] == 1:
        print(f'Outnput: {N} , location {pos}: wrong')
        record[pos]-=1
      else:
        N[pos] = 1
        print(f'Output: {N} , location {pos}: exact')
        if 0 in N:
          record[pos]+=10
          break
        else:
          record[pos]+=1
```
