Câu 1: Sử dụng thư viện Numpy, hãy tạo một vector a có 7 chiều (tương ứng 7 phần tử) với giá
trị các phần tử được sinh bằng hàm (kết quả xuất ra màn hình):
- np.arrange()
- np.linspace()
- np.logspace()
- np.random.rand()

Ghi chú: sau mỗi lần sinh ra giá trị cho vector, hãy cho biết số chiều của vector được sinh ra và
số chiều của vector chuyển vị của vector đó; cho biết kích thước của vector a và vector chuyển vị
của vector a.

Gợi ý: tham khảo ndim và shape.

```python
import numpy as np

a = np.arange(0, 7, 1)
print(a)
print(a.ndim)
print(a.shape)
```

```python
a = np.linspace(0, 10, 7)
print(a)
print(a.ndim)
print(a.shape)
```

```python
a = np.logspace(2, 3, 7)
print(a)
print(a.ndim)
print(a.shape)
```

```python
a = np.random.rand(7)
print(a)
print(a.ndim)
print(a.shape)
```

Câu 2: Sử dụng thư viện Numpy, hãy tạo ra ma trận A10x7 với giá trị các phần tử của ma trận A
được sinh bởi:
- np.arrange()
- np.linspace()
- np.logspace()
- np.random.rand()

Sau mỗi lần ma trận được tạo ra, hãy xuất kết quả ra màn hình và cho biết:
- Số chiều và kích thước của ma trận A và ma trận AT

```python
A = np.arange(1, 11).reshape(-1, 1) * np.arange(1, 8)

print(A)
print(A.size)
print(A.ndim)
print(A.shape)

print(A.T)
print(A.T.size)
print(A.T.ndim)
print(A.T.shape)
```

```python
A = np.linspace(1, 10, 10).reshape(-1, 1) * np.linspace(1, 10, 7)

print(A)
print(A.size)
print(A.ndim)
print(A.shape)

print(A.T)
print(A.T.size)
print(A.T.ndim)
print(A.T.shape)
```

```python
A = np.logspace(1, 2, 10).reshape(-1, 1) * np.logspace(1, 2, 7)

print(A)
print(A.size)
print(A.ndim)
print(A.shape)

print(A.T)
print(A.T.size)
print(A.T.ndim)
print(A.T.shape)
```

```python
A = np.random.rand(10, 7)

print(A)
print(A.size)
print(A.ndim)
print(A.shape)

print(A.T)
print(A.T.size)
print(A.T.ndim)
print(A.T.shape)
```

Câu 5: Thực hiện các yêu cầu sau:
- Xây dựng hàm tính giá trị trung bình
- Xây dựng hàm tính giá trị trung vị
- Xây dựng hàm tính giá trị yếu vị
- Xây dựng hàm tính phương sai
- Xây dựng hàm tính độ lệch chuẩn
- Hãy tạo ra ma trận A15x10 với các giá trị sinh bởi hàm arrange()
- Áp dụng 5 hàm ở trên cho từng hàng của ma trận A
- Áp dụng 5 hàm ở trên cho từng cột của ma trận A
Gợi ý: tham khảo hàm apply_along_axis() đã học ở bài trước.

```python
from scipy import stats

def mean(vector):
  return np.mean(vector)

def median(vector):
  return np.median(vector)

def mode(vector):
  return stats.mode(vector)

def var(vector):
  return np.var(vector)

def std(vector):
  return np.std(vector)

A = np.arange(1, 16).reshape(-1, 1) * np.arange(1, 11)
print(A)
print("ROW:")
print("mean:", np.apply_along_axis(mean, 1, A))
print("median:", np.apply_along_axis(median, 1, A))
print("mode:", np.apply_along_axis(mode, 1, A))
print("var:", np.apply_along_axis(var, 1, A))
print("std:", np.apply_along_axis(std, 1, A))
print("COLUMM:")
print("mean:", np.apply_along_axis(mean, 0, A))
print("median:", np.apply_along_axis(median, 0, A))
print("mode:", np.apply_along_axis(mode, 0, A))
print("var:", np.apply_along_axis(var, 0, A))
print("std:", np.apply_along_axis(std, 0, A))
```
