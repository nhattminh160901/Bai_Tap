```python
import numpy as np
import pandas as pd
```

```python
class game():
  def __init__(self, x_tom, y_tom, x_jerry, y_jerry, x_spike, y_spike, x_block, y_block): # tao doi tuong
    self.tom = (x_tom, y_tom)
    self.jerry = (x_jerry, y_jerry)
    self.spike = (x_spike, y_spike)
    self.block = (x_block, y_block)

  def set_V(self, x, y): # tao V
    self.V = np.zeros((x, y))
    self.V[self.block] = -100
    self.V[self.jerry] = 1
    self.V[self.spike] = -1
    self.df = pd.DataFrame(self.V)

  def action_tom(self, x_tom, y_tom):
    if (x_tom, y_tom) == self.block:
      print("dung tuong")
    elif (x_tom, y_tom) == self.spike:
      print("bi cho can")
    elif (x_tom, y_tom) == self.jerry:
      print("bat duoc con chuot")
      self.tom = (x_tom, y_tom)
    else:
      self.tom = (x_tom, y_tom)

  def display(self):
    self.showframe = pd.DataFrame(self.V)
    for _ in range(self.V.shape[0]):
      for __ in range(self.V.shape[1]):
        self.showframe.at[(_,__)]=" "

    self.showframe.at[self.jerry] = "J"
    self.showframe.at[self.spike] = "S"
    self.showframe.at[self.block] = "--"
    if self.tom == self.jerry:
      self.showframe.at[self.tom] = "w"
    else:
      self.showframe.at[self.tom] = "T"
    display(self.showframe)

  def get_pos_tom(self):
    print(self.tom)

  def move(self, x, y):
    up       = (x-1, y)
    down     = (x+1, y)
    left     = (x, y-1)
    right    = (x, y+1)
    moves    = [up, down, left, right]
    canMoves = []
    for move in moves:
        move_x, move_y = move
        if 0 <= move_x < self.V.shape[0] and 0 <= move_y < self.V.shape[1]:
              if move not in [self.block]:
                canMoves.append(move)
    return canMoves

  def allMoves(self):
    results = []
    for i in range(self.V.shape[0]):
      for j in range(self.V.shape[1]):
        obj = {(i, j): self.move(i, j)}
        results.append(obj)
    return results

  def addValue(self, value):
    values = []
    for _ in value:
        values.append(_)
    return values

  def updateValue(self, value, gamma, p_down):
    listValue = self.addValue(value)
    maxValue = 0
    if len(value) > 1:
        sumVal = len(value) - 1
    else:
        sumVal = len(value)
    result = []
    for i in range(len(listValue)):
        sum = 0
        x_i, y_i = listValue[i]
        for j in range(len(listValue)):
            x_j, y_j = listValue[j]
            if j == i:
                sum += 0.8 * gamma * self.V[x_i][y_i]
            else:
                sum += (p_down / sumVal) * gamma * self.V[x_j][y_j]
        result.append(sum)
    maxValue = max(result)
    return maxValue

  def qtable(self, gamma, p_down):
    for __ in range(30):
      for i in self.allMoves():
        for _ in i:
          if _ in (self.jerry, self.spike, self.block):
            continue
          value = i[_]
          self.V[_] = self.updateValue(value, gamma, p_down)
    return self.V

  def auto_move(self):
    up       = (self.tom[0]-1, self.tom[1])
    down     = (self.tom[0]+1, self.tom[1])
    left     = (self.tom[0], self.tom[1]-1)
    right    = (self.tom[0], self.tom[1]+1)

    results = self.move(self.tom[0], self.tom[1])
    arr = []
    for result in results:
      obj = {}
      obj = {"pos":result, "values":self.qtable()[result]}
      arr.append(obj)

    x, y = max(arr, key=lambda x:x['values'])['pos']
    if (x, y) == up:
      print("Di len")
      self.action_tom(x, y)
      self.get_pos_tom()
      self.display()
    elif (x, y) == down:
      print("Di xuong")
      self.action_tom(x, y)
      self.get_pos_tom()
      self.display()
    elif (x, y) == left:
      print("Qua trai")
      self.action_tom(x, y)
      self.get_pos_tom()
      self.display()
    elif (x, y) == right:
      print("Qua phai")
      self.action_tom(x, y)
      self.get_pos_tom()
      self.display()

    if (x, y) == self.jerry:
      return None

    self.auto_move()
    return None
```

```python
def canMove(x, y, V, block_moves):
  up       = (x-1, y)
  down     = (x+1, y)
  left     = (x, y-1)
  right    = (x, y+1)
  moves    = [up, down, left, right]
  canMoves = []
  for move in moves:
      move_x, move_y = move
      if 0 <= move_x < V.shape[0] and 0 <= move_y < V.shape[1]:
            if move not in block_moves:
              canMoves.append(move)
  return canMoves
```

```python
def allMoves(V, block_moves):
  results = []
  for i in range(V.shape[0]):
    for j in range(V.shape[1]):
      obj = {(i, j): canMove(i, j, V, block_moves)}
      results.append(obj)
  return results;
```

```python
def nextStep(x, y):

  nextMove = ''
  up       = (x-1, y)
  down     = (x+1, y)
  left     = (x, y-1)
  right    = (x, y+1)

  listMoveNexts = canMove(x, y, V, lock)
  arr = []

  for listMoveNext in listMoveNexts:
    obj = {}
    obj = {"pos":listMoveNext, "values":V[listMoveNext]}
    arr.append(obj)

  maxKey = max(arr, key=lambda x:x['values'])['pos']

  if maxKey == up:
      next_move = '^'
  elif maxKey == down:
      next_move = 'v'
  elif maxKey == left:
      next_move = '<'
  elif maxKey == right:
      next_move = '>'

  df.at[x, y] = next_move

  i, j = maxKey

  if maxKey == (0, 3):
      return df
  nextStep(i, j, V, df)

  return df
```
